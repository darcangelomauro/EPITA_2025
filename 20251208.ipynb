## Practical impelementation topics

### Register layouts

```python
from pulser import Register

reg = Register({
    "q1": [0,0],
    "q2": [0.35235, 9.234523]
})

reg.draw()
```

```python
# Things on the machine are a bit more complicated.
# need to use layouts

from pulser.register.special_layouts import TriangularLatticeLayout

layout = TriangularLatticeLayout(60, 5)
layout.draw()
```

```python
# to define a register out of a layout, select the trap ids
# that compose your register

trap_ids = [25, 34, 29]
reg = layout.define_register(*trap_ids)
reg.draw()
```

### Analog device

```python
#So far we've always used MockDevice, which is convenient for
#prototyping an algorithm. but eventually, in a real implementation,
#we will need to make sure that the algo can run on a real machine.
#In this case we can swap MockDevice for a device with more realistic characteristics,
#called AnalogDevice
```

```python
from pulser.devices import AnalogDevice, MockDevice

A = AnalogDevice
M = MockDevice

A.print_specs()
M.print_specs()
```
